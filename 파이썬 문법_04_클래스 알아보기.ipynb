## 4-4 사용자 클래스로 객체 속성을 넣기 

- 객체 내의 속성을 __init__ 메소드 내부에 할당한다. 

### 객체를 생성할 때는 초기화 함수에 정의한 매개변수에 해당하는 인자를 넣는다

```python
Image("./이미지/클래스초기화함수.gif", width=400, height=400)
```

###  클래스 내부에 객체 의 속성을 정의한다.
- self 를 넣는 이유는 객체를 지정해서 그 내부에 속성을 생성한다는 뜻이다

```python
class Person :
    def __init__(self, name) :
        self.name = name
        
```

#### 객체를 만들때는 __init__ 메소드의 self를 제외한 매개변수와 동일한 인자를 전달한다

```python
p_ = Person("클래스")
```

```python
p_
```

```python
p_.name
```

```python
p_.__dict__
```

## 4-5 사용자 클래스의 메소드 넣기
- 객체가 사용할 수 있는 메소드를 정의할 때 첫번째 매개변수는 항상 self를 넣는다.
- 실제 객체가 self 자리에 들어와서 메소드를 처리한다.

### 클래스를 정의할 때 두 개의 속성과 두 개의 메소드를 정의한다. 

- 메소드는 단순하게 조회만 할 수 있도록 내부에 매개변수는 self 만 넣는다

```python
class Person :
    def __init__(self, name, age) :
        self.name = name
        self.age = age
        
    def getName(self) :
        print(" getName 실행 ")
        return self.name
    
    def getAge(self) :
        print(" getAge 실행")
        return self.age
```

#### 객체를 생성한다

```python
p_1 = Person("이름", 24)
```

### 메소드 호출 방법

- 객체 다음에 점연산자와 메소드 이름을 작성한다.
- 메소드를 실행 연산자 안에 인자를 전달한다.
- self는 객체가 자동으로 세팅되므로 아무런 인자도 전달할 필요가 없다

```python
Image("./이미지/메소드호출방법.png", width=500, height=300)
```

#### 두 개의 메소드를 실행한다

```python
p_1.getName()
```

```python
p_1.getAge()
```

#### 파이썬은 속성을 바로 사용할 수 있다

- 파이썬은 다른 프로그램 언어와 달리 객체의 속성을 바로 접근해서 사용할 수 있다.


```python
p_1.name
```
